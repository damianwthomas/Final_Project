```python
import pandas as pd
import numpy as np
```

```python
nba_data = pd.read_csv("Resources/nba_data.csv")
nba_data.head()
```

```python
category = {'Y':1,'N':0}
nba_data.hof_flag = [category[item] for item in nba_data.hof_flag] 
nba_data.final_mvp = [category[item] for item in nba_data.final_mvp] 
nba_data.all_stargame = [category[item] for item in nba_data.all_stargame] 
nba_data.championship_ring = [category[item] for item in nba_data.championship_ring]
nba_data.league_mvp = [category[item] for item in nba_data.league_mvp] 
```

```python
nba_data.head()
```

```python
nba_data.describe()
```

```python
features = nba_data [['final_mvp', 'all_stargame','championship_ring','league_mvp','assist', 'blocks', 'box_plus_minus','points','steals','total_rebounds','value_of_replacement','win_shares']].values
classes = nba_data ['hof_flag'].values
feature_names = ['final_mvp', 'all_stargame','championship_ring','league_mvp','assist', 'blocks', 'box_plus_minus','points','steals','total_rebounds','value_of_replacement','win_shares']

features
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(1234)
(training_inputs,
 testing_inputs,
 training_classes,
 testing_classes) = train_test_split(features, classes, train_size = 0.75, random_state=1)
```

```python
testing_inputs.shape
```

```python
training_inputs.shape
```

```python

#Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=1)

clf.fit(training_inputs, training_classes)
```

```python
from IPython.display import Image
from sklearn.externals.six import StringIO
from sklearn import tree
from pydotplus import graph_from_dot_data


dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data,
                        feature_names=feature_names)
graph = graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
clf.score(training_inputs, training_classes)
```

```python
clf.score(testing_inputs, testing_classes)
```

```python
# cross val Fold validation (10 differecnt folds of the data)
```

```python
from sklearn.model_selection import cross_val_score

clf = DecisionTreeClassifier(random_state=1)
cv_scores =  cross_val_score(clf, features, classes, cv=10)

cv_scores.mean()
```

```python
# Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=10, random_state=1)
cv_scores =  cross_val_score(clf, features, classes, cv=10)

cv_scores.mean()
```

```python
# KNN
```

```python
from sklearn import neighbors

clf = neighbors.KNeighborsClassifier(n_neighbors=10)
cv_scores = cross_val_score(clf, features, classes, cv=10)

cv_scores.mean()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.linear_model import LogisticRegressionCV

clf = LogisticRegressionCV(random_state=1, max_iter=1000, cv=10).fit(training_inputs, training_classes)
#cv_scores = cross_val_score(clf, features, classes, cv=10)
#cv_scores.mean()
```

```python
clf
```

```python
clf.score(testing_inputs, testing_classes)
```

```python
hof_predict = clf.predict(testing_inputs)
```
